```python
import pandas as pd
import nltk
import re
import numpy as np
from nltk.corpus import stopwords
import tensorflow as tf
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import one_hot
```

```python
df = pd.read_csv("/content/drive/MyDrive/NLP/AI ML Python/Tweet_NFT.xlsx - Sheet1.csv")
df.head()
```

```python
df['tweet_intent'].value_counts()
```

```python
df.isnull().sum()
```

```python
st = 127453-31089
train_df = df.iloc[:st,:]
test_df = df.iloc[st:,:]
train_df.shape,test_df.shape
```

```python
train_df.isnull().sum()
```

```python
nltk.download('stopwords')
```

```python
### Data preprocessing
voca_size = 50000
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0,len(train_df)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ' , train_df['tweet_text'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_rep = [one_hot(words,voca_size) for words in corpus]
```

```python
onehot_rep
```

```python
sen_length = 50
embadding_doc = pad_sequences(onehot_rep,padding='pre', maxlen=sen_length)
print(embadding_doc)
```

```python
train_df['tweet_intent'].unique()
```

```python
result = {'Community':2, 'Giveaway':4, 'Appreciation':1, 'Presale':7, 'Whitelist':9,
       'pinksale':8, 'Done':3, 'Interested':5, 'Launching Soon':6}
```

```python
train_df['tweet_intent'] = train_df['tweet_intent'].map(result)
```

```python
train_df['tweet_intent']
```

```python
X_final = np.array(embadding_doc)
Y_final = np.array(train_df['tweet_intent'])
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,Y_train,y_test = train_test_split(X_final,Y_final,test_size=0.25,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
rf = RandomForestClassifier(n_estimators=500, max_depth=12, n_jobs=-1)
rf_model = rf.fit(X_train, Y_train)
```

```python
y_pred_prob = rf_model.predict_proba(x_test)
print(y_pred_prob)
```

```python
y_pred = rf_model.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred)) 
print('---------------------------------------------------')
print(classification_report(y_test,y_pred))  
print('---------------------------------------------------')
print(accuracy_score(y_test, y_pred)) 
```

```python
from xgboost import XGBClassifier
mdl_xgb = XGBClassifier(n_estimators = 100, learning_rate = 0.1, max_depth =10)
mdl_xgb.fit(X_train, Y_train)
y_pred = mdl_xgb.predict(x_test)

print(confusion_matrix(y_test,y_pred)) 
print('---------------------------------------------------')
print(classification_report(y_test,y_pred))  
print('---------------------------------------------------')
print(accuracy_score(y_test, y_pred)) 
```
